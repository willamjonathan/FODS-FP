```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Fitting Random Forest Regression to the dataset
# import the regressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
#reading the data from the csv file
df = pd.read_csv('FinalResult.csv')
```

```python
df['Index'] = 'Big'
df.loc[df['Alive (%)']<= 96	,'Index'] ="Small" 
X= df [['ART (%)', 'People with HIV']]
y= df ['Index']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier( random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV

parameter_grid = { 
    'n_estimators': [200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    
    'criterion' :['gini', 'entropy']
}

grid_search = GridSearchCV(estimator=rf_model, cv = 5, param_grid= parameter_grid)
grid_search.fit(X_train, y_train)
```

```python
pd.DataFrame(grid_search.cv_results_)
```

```python
final_model = grid_search.best_estimator_
final_model
```

```python
model_train = RandomForestClassifier(n_estimators = 500)
model_train.fit(X_train, y_train)
```

```python
# make prediction
predictions = model_train.predict(X_test)
predictions
```

```python
# Function to calculate accuracy
def Report(y_test, y_pred):
    print("Report : ",
    classification_report(y_test, y_pred))
```

```python
print("----------------------Random Forest---------------------------")
print("Trained acc: ", accuracy_score(y_train, model_train.predict(X_train)))
print("Test accuracy:", accuracy_score(y_test, predictions))
print("Confusion matrix train: ", "\n", confusion_matrix(y_train, model_train.predict(X_train)))
print("Confusion matrix test: ", "\n",confusion_matrix(y_test, predictions))
print(Report(y_train, model_train.predict(X_train)))
print("---------------------------------------------------------")
print(Report(y_test, predictions))
```
