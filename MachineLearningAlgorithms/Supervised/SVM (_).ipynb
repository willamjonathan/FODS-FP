```python
S#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from sklearn import datasets
from sklearn.svm import  SVC
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
#reading the data from the csv file
df = pd.read_csv('FinalResult.csv')
df.info()
df.head()
```

```python
df['Index'] = 'Big'
df.loc[df['Alive (%)']<= 96	,'Index'] ="Small" 
X= df [['ART (%)', 'People with HIV']]
y= df ['Index']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
svm_model = SVC( random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV

parameter_grid = { 'C': [0.1, 1, 10, 100, 1000],
                    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
                'kernel': ['rbf']
}

grid_search = GridSearchCV(estimator=svm_model, cv = 3, param_grid= parameter_grid)
grid_search.fit(X_train, y_train)
```

```python
pd.DataFrame(grid_search.cv_results_)
```

```python
final_model = grid_search.best_estimator_
final_model
```

```python
svm_class =SVC(C=100, gamma=0.0001, kernel = "rbf",random_state=42)
svm_class.fit(X_train, y_train)
predictions = svm_class.predict(X_test)
predictions
```

```python
# Function to calculate accuracy
def Report(y_test, y_pred):
    print("Report : ",
    classification_report(y_test, y_pred))
```

```python
print("-----------------------SVM-----------------------------")
print("Trained acc: ", accuracy_score(y_train, svm_class.predict(X_train)))
print("Test accuracy:", accuracy_score(y_test, predictions))
print("Confusion matrix train: ", "\n", confusion_matrix(y_train, svm_class.predict(X_train)))
print("Confusion matrix test: ", "\n",confusion_matrix(y_test, predictions))
from sklearn.metrics import classification_report
print(Report(y_train, svm_class.predict(X_train)))
print("---------------------------------------------------------")
print(Report(y_test, predictions))
```
